This is the "scratchpad" we used to understand Object Oriented Programming before building the Smart Home.

* **Class:** The Blueprint (`SimpleDevice`)
* **Object:** The physical thing created from the blueprint (`kitchen_light`)

## The Blueprint (Class)

```python
class SimpleDevice:
    def __init__(self):
        self.is_on = False
    def flip_switch(self):
        self.is_on = not self.is_on
        if self.is_on:
            print("The Device is now on")
        else:
            print("The Device is now off")
```

## The Object (Real Device)

```python
kitchen_light = SimpleDevice()
kitchen_light.flip_switch()
```
